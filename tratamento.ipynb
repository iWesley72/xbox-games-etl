<p>Vamos consertar valores estranhos no dataset, como as colunas NOME e DATA_LANCAMENTO.</p>

## Bibliotecas e dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dados.csv', sep=';')
df
```

## Coluna NOME

```python
df['NOME']
```

```python
nomeCorrigido = []
for nome in df['NOME']:
    nome = nome.split('>')
    nomeCorrigido.append(nome[3].replace('-', ' '))
print(nomeCorrigido)
    
```

```python
dfTratado = df
dfTratado['NOME'] = nomeCorrigido
dfTratado
```

## Coluna DATA_LANCAMENTO

```python
datasCorrigidas = []
for data in df['DATA_LANCAMENTO']:
    data = data.split('T')
    datasCorrigidas.append(data[0])
print(datasCorrigidas)
```

```python
dfTratado['DATA_LANCAMENTO'] = datasCorrigidas
dfTratado
```

```python
datasCorrigidas = pd.to_datetime(dfTratado['DATA_LANCAMENTO'], format='%Y-%m-%d', errors='coerce')
```

```python
dfTratado[dfTratado['DATA_LANCAMENTO'] == '0']
```

```python
dfTratado['DATA_LANCAMENTO'] = datasCorrigidas
dfTratado.info()
```

## Arquivo csv tratado

```python
dfTratado.to_csv('dadosTratado.csv', sep=';')
```
